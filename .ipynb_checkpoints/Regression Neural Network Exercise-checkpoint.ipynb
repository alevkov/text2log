```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras.wrappers.scikit_learn import KerasRegressor
```

# Load and visualize data set

```python
dataset = np.loadtxt("data.csv", delimiter=",")
dataset
```

```python
# Split independent from dependent vars
X = dataset[:,1:5]
y = dataset[:,5]
```

```python
X # Dependent variables
```

# Preprocessing

```python
y = np.reshape(y, (-1,1))

# We must scale the values in the range [0, 1] in order for Stochastic Gradient Descent to take more measured steps
scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()

# MinMax everything except the "Process"
scaler_x.fit(X[:,0:3])
xscale=scaler_x.transform(X[:,0:3])
scaler_y.fit(y)
yscale=scaler_y.transform(y)
 
# Append "Process" to the dataset
xscale = np.c_[xscale, X[:,3]]
```

```python
xscale
```

```python
# The train/test split that yielded the best performance with this dataset was 75/25 (default)
X_train, X_test, y_train, y_test = train_test_split(xscale, yscale, test_size=0.2)
```

# Model visualization

```python
model = Sequential()
model.add(Dense(12, input_dim=4, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(1, activation='linear'))
model.summary()
```

# Hyperparameter selection

```python
# Stochastic Gradient Descent is chosen as the optimizer function
# compile model
opt = tf.keras.optimizers.SGD(lr=0.01, momentum=0.9)
model.compile(loss='mse', optimizer=opt, metrics=['mse'])
```

# Training

```python
history = model.fit(X_train, y_train, epochs=200, batch_size=1, verbose=1)
```

# Plot showing loss decay

```python
plt.plot(history.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['MSE'], loc='upper left')
plt.show()
```

```python
model.evaluate(X_test, y_test)
```
